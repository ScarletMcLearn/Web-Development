```python
from bs4 import BeautifulSoup

with open("Test Files/index.html") as fp:
    soup = BeautifulSoup(fp, 'lxml')
```

```python
href_tags = soup.find_all(href=True)
src_tags = soup.find_all(src=True)
```

```python
site_links = []
static_links = []
garb_links = []

for _ in href_tags:
    if((_['href'].split(':')[0]== 'http') or (_['href'].split(':')[0]== 'https')):
        site_links.append((_['href']))
    elif(_['href'].rsplit('.', 1)[-1] != 'html'):
        static_links.append(_['href'])
    else:
        garb_links.append(_['href'])

site_links_2 = []
static_links_2 = []

for _ in src_tags:
    if((_['src'].split(':')[0]== 'http') or (_['src'].split(':')[0]== 'https')):
        site_links_2.append((_['src']))
    elif(_['src'].rsplit('.', 1)[-1] != 'html'):
        static_links_2.append(_['src'])
    else:
        garb_links.append(_['src'])
```

```python
full_static_link = static_links + static_links_2
cleaned_static_list = []

for _ in full_static_link:
    if (_.split('/')[0] == 'img') or (_.split('/')[0] == 'js') or (_.split('/')[0] == 'css'):
        print(_)
        cleaned_static_list.append(_)
```

```python
import re
```

```python
indx_lst = []

for _ in cleaned_static_list:
    for m in re.finditer(_, str(soup)):
        indx_lst.append([_, m.start(), m.end()])
```

```python
f_strt = "\"{% static \'"
f_end = "\' %}\""

html_doc = str(soup).replace("\'", '\"')

for itm in cleaned_static_list:
    html_doc = html_doc.replace('"' + itm + '"', f_strt + itm + f_end)
    
print("Done!")
# cleaned_static_list
```
