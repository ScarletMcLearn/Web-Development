```python
import time

with open('cleaned_data_url.txt', 'r') as f:
    data = f.readlines()
    
    for j in range(len(data)):
        source = data[j]

    #     print(source)

        current_page = web_scraper.soupified_request(source)

        tag_text_list = []

        title = ArticleScraper.get_article_title(current_page)

        body = list(current_page.find('div', {'class' : 'field-item even first last'}))

        ArticleScraper.tag_text_list_gen(body, tag_text_list)

        contentss = []

        for i in range(1, len(tag_text_list)):

            tagg = tag_text_list[i][0]
            textt = tag_text_list[i][1]

            contentss.append({'tag':tagg, 'content':textt})


    #     print(title)

    #     print(contentss)


    #     print(contentss)

    #         article_1 = AMA_Article(title=title, contents=contentss, source_url=line)

    #         article_1.save()

        print('\nCompleted URL...\n')

        print('\nSleeping for 11 secs!\n')
        time.sleep(11)
    #        
        j_data = {'title': title, 'source':source.strip(), 'contents':contentss}

        with open(('AMA Resources/' + str(j)), 'w') as output_file:
            json.dump(j_data, output_file)

```
